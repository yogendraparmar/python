```python
x = df.drop(columns=['Survived'])
y= df['Survived']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
x_train.sample(8)
```

```python
#create new columne replace (99 and -1)
x_train['Age_99']= x_train['Age'].fillna(99)
x_train['Age_minuse1']= x_train['Age'].fillna(-1)
x_train['Fare__99']= x_train['Fare'].fillna(99)
x_train['Fare_minuse1']= x_train['Fare'].fillna(-1)
```

```python
x_train.sample(10)
```

```python
#variance review
print('Orginal age variance review:',x_train['Age'].var())
print('Age after imputatonn 99 value variation:',x_train['Age_99'].var())
print('Age after imputation -1 value variable:',x_train['Age_minuse1'].var())

print('Orginal fare variance review:',x_train['Fare'].var())
print('Age after imputatonn 99 value variation:',x_train['Fare__99'].var())
print('Age after imputation -1 value variable:',x_train['Fare_minuse1'].var())
```
